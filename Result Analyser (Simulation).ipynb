```python
# Imports
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define path
path = "/Users/v.d bosch/Documents/Github/Score-driven_Tensor_Factor_Model/"
```

```python
# Define method plotting the boxplot
def show_boxplot(data_a,data_b,ticks,label_a,label_b):

    # Set the colors
    def set_box_color(bp, color):
        plt.setp(bp['boxes'], color=color)
        plt.setp(bp['whiskers'], color=color)
        plt.setp(bp['caps'], color=color)
        plt.setp(bp['medians'], color=color)

    plt.figure(figsize=(7, 4))
    plt.grid(which='major', axis='y', color='#DDDDDD', linestyle=':', linewidth=1.0)

    bpl = plt.boxplot(data_a, positions=np.array(range(len(data_a)))*2.0-0.3, widths=0.5,flierprops={'marker': 'o', 'markersize': 5, 'mec': 'navy'})
    bpr = plt.boxplot(data_b, positions=np.array(range(len(data_b)))*2.0+0.3, widths=0.5,flierprops={'marker': 'o', 'markersize': 5, 'mec': 'orangered'})
    set_box_color(bpl, 'navy')
    set_box_color(bpr, 'orangered')

    plt.plot([], c='navy', label=label_a)
    plt.plot([], c='orangered', label=label_b)
    plt.legend(bbox_to_anchor=(0, 1.02, 1, 0.2), loc="lower left", mode="expand", borderaxespad=0, ncol=2)

    plt.xticks(range(0, len(ticks) * 2, 2), ticks)
    plt.xlim(-1, len(ticks)*2-1)
    plt.ylim(-0.08, 1.08)
    plt.tight_layout()
    plt.show()
```

```python
# Read in all data
df_1_5 = pd.read_excel(path+"Simulation_Results_K=1_d1=5_Check.xlsx",index_col=0)
df_2_5 = pd.read_excel(path+"Simulation_Results_K=2_d1=5_Check.xlsx",index_col=0)
df_3_5 = pd.read_excel(path+"Simulation_Results_K=3_d1=5_Check.xlsx",index_col=0)


# Show all 4 boxplots (RMSE, factors)
show_boxplot(data_a=[df_1_5.iloc[:,0], df_2_5.iloc[:,0], df_3_5.iloc[:,0]],
             data_b=[df_1_5.iloc[:,1], df_2_5.iloc[:,1], df_3_5.iloc[:,1]],
             ticks=['K=1', 'K=2', 'K=3'],
             label_a='RMSE first factor',
             label_b='Average RMSE all factors')

# Show all 4 boxplots (RMSE, loadings)
show_boxplot(data_a=[df_1_5.iloc[:,2], df_2_5.iloc[:,2], df_3_5.iloc[:,2]],
             data_b=[df_1_5.iloc[:,3], df_2_5.iloc[:,3], df_3_5.iloc[:,3]],
             ticks=['K=1', 'K=2', 'K=3'],
             label_a='RMSE first loading',
             label_b='Average RMSE all loadings')

# Show all 4 boxplots (MAE, factors)
show_boxplot(data_a=[df_1_5.iloc[:,4], df_2_5.iloc[:,4], df_3_5.iloc[:,4]],
             data_b=[df_1_5.iloc[:,5], df_2_5.iloc[:,5], df_3_5.iloc[:,5]],
             ticks=['K=1', 'K=2', 'K=3'],
             label_a='MAE first factor',
             label_b='Average RMSE all factors')

# Show all 4 boxplots (MAE, loadings)
show_boxplot(data_a=[df_1_5.iloc[:,6], df_2_5.iloc[:,6], df_3_5.iloc[:,6]],
             data_b=[df_1_5.iloc[:,7], df_2_5.iloc[:,7], df_3_5.iloc[:,7]],
             ticks=['K=1', 'K=2', 'K=3'],
             label_a='MAE first loading',
             label_b='Average RMSE all loadings')
```
